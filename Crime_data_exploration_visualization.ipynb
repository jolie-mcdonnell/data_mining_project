#** CRIME DATA**

# Dataset Description
Crime incident reports are provided by Boston Police Department (BPD) to document the initial details surrounding an incident to which BPD officers respond. This is a dataset containing records from the new crime incident report system, which includes a reduced set of fields focused on capturing the type of incident as well as when and where it occurred. Records in the new system begin in June of 2015.

The data is organized on data.boston.gov by year and each year's dataset is loaded separately into the drive folder crimes as "crime_20**.csv"

# Description of Fields
*   INCIDENT_NUMBER: internal BPD report number
*   OFFENSE_CODE: numerical code of offense description
*   OFFENSE_CODE_GROUP: internal categorization of offense description
*   OFFENSE_DESCRIPTION: primary descriptor of incident
*   DISTRICT: what district the crime was reported in
*   REPORTING_AREA: RA number associated with where the crime was reported from
*   SHOOTING: indicated a shooting took place
*   OCCURRED_ON_DATE: earliest date and time the incident could have taken place
*   YEAR
*   MONTH
*   DAY_OF_WEEK
*   HOUR
*   UCR_PART: universal crime reporting part number (1, 2, 3)
*   STREET: street name the incident took place
*   Lat: latitude, 0.0 if null
*   Long: longitude, 0.0 if null
*   Location: tuple (lat, long)

```python
# 2020 Crime Data
crime20 = pd.read_csv("/content/drive/MyDrive/dm_project_data/crime/crime_2020.csv")
crime20.head()
```

```python
# Plot showing number of crimes per district
crime20['DISTRICT'].value_counts()[:20].plot(kind='barh')
```

```python
# Total counts of types of offense
crime20['OFFENSE_DESCRIPTION'].value_counts()[:20].plot(kind='barh')
```

```python
# Looking at types of crimes in the district with the most crimes vs the district with the least crimes
crime20B4 = crime20[crime20['DISTRICT'] == 'B2']
crime20B4['OFFENSE_DESCRIPTION'].value_counts()[:20].plot(kind='barh')

```

```python
crime20B4 = crime20[crime20['DISTRICT'] == 'A15']
crime20B4['OFFENSE_DESCRIPTION'].value_counts()[:20].plot(kind='barh')
```

```python
# Scatterplot of 2020 crime locations

# remove instances where lat and long are null (0)
crime20clean = crime20[crime20['Lat'] > 42]
crime20clean = crime20clean[crime20clean['Long'] < -70.75]

# Plot scatter to get idea of shape
plt.scatter(crime20clean['Lat'], crime20clean['Long'])
plt.xlim(42.23, 42.4)
plt.ylim(-71, -71.2)
```

```python
# More complex mapping of 2020 crime regions
# https://towardsdatascience.com/easy-steps-to-plot-geographic-data-on-a-map-python-11217859a2db

BBox = ((crime20clean.Long.min(),crime20clean.Long.max(),crime20clean.Lat.min(),crime20clean.Lat.max()))

map = plt.imread('/content/drive/MyDrive/dm_project_data/map.png')
fig, ax = plt.subplots(figsize = (8,7))
ax.scatter(crime20clean.Long, crime20clean.Lat, zorder=1, alpha= 0.1, c='b', s=2)
ax.set_title('Spatial Plot of Crimes in Boston')
ax.set_xlim(BBox[0],BBox[1])
ax.set_ylim(BBox[2],BBox[3])
ax.imshow(map, zorder=0, extent = BBox, aspect= 'equal')
```
